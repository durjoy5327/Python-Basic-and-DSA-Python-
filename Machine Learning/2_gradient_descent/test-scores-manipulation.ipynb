```python
df= pd.read_csv('/kaggle/input/test-scores/test_scores.csv')
df
```

```python
x= np.array(df.math.values)
y= np.array(df.cs.values)
x,y
```

```python
def gradient_descent(x, y, learning_rate=0.01):
    m_cur = b_cur = 0 
    n = len(x)
    previous_cost = float('inf')
    iterations = 50000

    for i in range(iterations):
        y_predicted = m_cur * x + b_cur  # Predicted values
        cost = (1 / n) * sum([val**2 for val in (y - y_predicted)])

        if math.isclose(cost, previous_cost, rel_tol=1e-20):
            print(f"Converged at iteration {i} with cost {cost:.4f}")
            break

        previous_cost = cost

        md = (-2 / n) * sum(x * (y - y_predicted))  
        bd = (-2 / n) * sum(y - y_predicted)        
        m_cur = m_cur - learning_rate * md          
        b_cur = b_cur - learning_rate * bd   

        plt.plot(x, y_predicted, color='green', alpha=0.3)

        print(f"m = {m_cur:.4f}, b = {b_cur:.4f}, cost = {cost:.4f}, iteration = {i}")

    y_final = m_cur * x + b_cur
    plt.plot(x, y_final, color='blue')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent')
    plt.show()
```

```python
gradient_descent(x,y, 0.00001)
```
